*GROUP BY*

the groupby() function in pandas is used to group data based on one or more columns in a Dataframe.

```python
s_qty = df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)

sns.barplot(x='Size', y='Qty', data = s_qty)
```

*Note: From the above Graph you can see that most of the Qty buys M-Size in the sales*

```python
# Checking B2B Dta by using pie chart

B2B_check = df['B2B'].value_counts()

# Plot the Pie Chart (Pls note: if you want to see True or False, enter .index after the labels)

plt.pie(B2B_check, labels= B2B_check.index, autopct='%1.1f%%')
plt.title('B2B Data')
#plt.axis ('equal')
plt.show()
```

*Note: from the above chart we can see the maximum i.e. 99.2% of buyers are retailer and 0.7% are B2B buyers*

```python
#Prepare data for scatter plot 

x_data = df['Category']
y_data = df['Size']

#Plot the scatter plot

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#Prepare data for scatter plot 

plt.figure(figsize=(12,6))
sns.countplot(data = df, x = 'ship-state')
plt.xlabel('Ship - State')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Top 10 states 

top_10_states = df['ship-state'].value_counts().head(10)

#plot count of citites by state 

plt.figure(figsize=(12,6))

sns.countplot(data = df[df['ship-state'].isin(top_10_states.index)], x='ship-state')
plt.xlabel('Ship - State')
plt.ylabel('Count')
plt.title('Distribution of State')
plt.xticks(rotation = 45)
plt.show()
```

*Note: From the above Graph you can see that most of the buyers are from Maharashtra State*

```python
# Get top 10 cities (already sorted in descending order by default)
top_10_city = df['ship-city'].value_counts().head(10)

# Sort the top 10 cities in ascending order
top_10_city = top_10_city.sort_values(ascending=False)

# Plot count of cities
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-city'].isin(top_10_city.index)], x='ship-city', order=top_10_city.index)

plt.xlabel('Ship - City')
plt.ylabel('Count')
plt.title('Distribution of Ship Cities')
plt.xticks(rotation=45)
plt.show()
```

*Note: From the above Graph you can see that most of the buyers are from Bengaluru City*
