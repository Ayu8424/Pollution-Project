```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/refs/heads/master/gas_prices.csv')
```

```python
gas
```

```python
gas.info()
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas Price Over the Year in USD')

plt.plot(gas.Year, gas.USA, 'b.-', label='United States')   #(b = color, . = markers, - = line)
plt.plot(gas.Year, gas.Canada,'r.-', label= 'Canada')
plt.plot(gas.Year, gas['South Korea'],'y.-', label= 'South Korea')

plt.xlabel('Years')
plt.ylabel('Prices in USD($)')

plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

If we want to see all the country line graph. Also in as fast as possible.

```python
plt.figure(figsize=(8,5))
plt.title('Gas Price Over the Year in USD')

# plt.plot(gas.Year, gas.USA, 'b.-', label='United States')   #(b = color, . = markers, - = line)
# plt.plot(gas.Year, gas.Canada,'r.-', label= 'Canada')
# plt.plot(gas.Year, gas['South Korea'],'y.-', label= 'South Korea')

for country in gas:
    if country != 'Year':
        plt.plot(gas.Year, gas[country], marker='.', label=[country])

plt.xlabel('Years')
plt.ylabel('Prices in USD($)')

plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

If we want to see specific country line graph. Also in as fast as possible.

```python
plt.figure(figsize=(8,5))
plt.title('Gas Price Over the Year in USD', fontdict={'fontweight':'bold', 'fontsize':18})

# plt.plot(gas.Year, gas.USA, 'b.-', label='United States')   #(b = color, . = markers, - = line)
# plt.plot(gas.Year, gas.Canada,'r.-', label= 'Canada')
# plt.plot(gas.Year, gas['South Korea'],'y.-', label= 'South Korea')

country_to_look_at = ['Australia', 'USA', 'Japan', 'Canada', 'UK']

for country in gas:
    if country in country_to_look_at:
        plt.plot(gas.Year, gas[country], marker='.', label=[country])

plt.xlabel('Years')
plt.ylabel('Prices in USD($)')

plt.xticks(gas.Year[::3])

# plt.xticks(gas.Year[::3].tolist()+[2011]) this is used to extend your list or x axis upto 2011. eventhough you have data or not. 

plt.legend()

# plt.savefig('Gas_price.png', dpi=300)

plt.show()
```
