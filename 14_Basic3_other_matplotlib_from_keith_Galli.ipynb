```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
fifa = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/refs/heads/master/fifa_data.csv')
```

```python
fifa.head(5)
```

HISTOGRAM

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins, color='yellow')
plt.xticks(bins)

plt.title('Distribution of Player Skill in FIFA', fontdict={'fontsize': 18, 'fontweight': 'bold'})
plt.xlabel('Skill level')
plt.ylabel('Number of Player')
plt.show() 
```

PIE CHARTS

```python
fifa.head(5)
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

right
```

```python

labels = ['left','right']
color = ['Yellow', 'Pink']

plt.pie([left,right], labels=labels, colors=color, autopct='%.2f %%')  #remember this '%.2f %%'.... this is use to show percentage.
plt.title('Foot Preference of Fifa Players', fontdict={'fontsize':18,'fontweight':'bold'})

plt.show()
```

Advanced PIE CHART

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight > 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight > 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight > 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight > 200].count()[0]

weight = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125', '125-150', '150-175','175 - 200', 'Above 200']
# color = ['Pink','Red','Yellow','Green','Orange']  instead of this we can also use 'ggplot'

plt.style.use('ggplot')

explode = (.4,0,0,0,.4)  #if there is a situation where 2 or more pct are in same place, we can use explode option. or pctdistance = 0.8 also

plt.pie(weight, labels=labels, autopct='%.2f %%', explode=explode)
plt.title('Weight of Player in FIFA (in lbs)', fontdict={'fontsize':18, 'fontweight':'bold'})


plt.show()
```

BOX and WHISKER PLOT

```python
fifa['Club'].unique()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
bayern = fifa.loc[fifa.Club == 'FC Bayern München']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'FC Bayern']

boxes = plt.boxplot([barcelona, madrid,bayern], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    box.set(color='#4286f4', linewidth=2) # Set edge color ... to execute this we have to mention medianprops={'linewidht':2}
    box.set(facecolor='#e0e0e0') #change fill color... to execute this we have to mention patch_artist = true

plt.title('Proffesional Soccer Team Comparision',fontdict={'fontsize':18,'fontweight':'bold'})
plt.ylabel('Fifa Overall Rating')

plt.show()
```
