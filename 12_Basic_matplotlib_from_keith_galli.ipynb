```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# plt.plot(x,y, label='2x', color = 'red', linewidth = 4, marker ='.', linestyle= '--', markersize = 14, markeredgecolor = 'blue') 
# linewidht for making the line more wider.  
# plt.title('Our First Graph!')

plt.plot(x,y, 'b^--', label='2x' )    # this is used to shortcut the samething which we did earlier making the big line. here we need 'b^--'
plt.title('Our First Graph!',  fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})  #fontdic is use for new font and fontsize for the size itself.

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])  # this is use to re size the x, y axis 
plt.yticks([0,2,4,6,8,10])
plt.legend()  #this is used to get legend in the graph itself, first you have to define the label then do the legend thing.
plt.show()
```

Resize your graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]



# to resize your graph
plt.figure(figsize=(5,3), dpi=100)

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label ='X^2')
plt.plot(x2[5:],x2[5:]**2, 'r--')



plt.plot(x,y, 'b^--', label='2x' )    # this is used to shortcut the samething which we did earlier making the big line. here we need 'b^--'
plt.title('Our First Graph!',  fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})  #fontdic is use for new font and fontsize for the size itself.

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])  # this is use to re size the x, y axis 
plt.yticks([0,2,4,6,8,10])
plt.legend()  #this is used to get legend in the graph itself, first you have to define the label then do the legend thing.

# plt.savefig('mygraph.png', dpi= 150) this is how you can save the graph
plt.show()
```

BAR CHART    

```python
label = ['A','B','C']
values = [2,8,4]
bars = plt.bar(label,values)

# bars[0].set_hatch('/')
# bars[1].set_hatch('O')   this can also be used or the next one can also be used to fill the column
# bars[2].set_hatch('*')

     # OR

patterns = ['/', 'O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4))
plt.show()
```
