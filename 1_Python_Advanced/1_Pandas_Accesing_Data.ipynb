```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[:, 'salary_rate': 'salary_hour_avg'].dropna(subset='salary_rate')
```

```python
median_year_salary = df.salary_year_avg.median()
median_year_salary
```

```python
median_hour_salary = df.salary_hour_avg.median()
median_hour_salary
```

```python
df_filled = df
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_year_salary)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_hour_salary)
```

```python
df_filled.loc[0:10, 'salary_year_avg':'salary_hour_avg']
```

```python
df_unique = df_filled
df_unique= df_unique.drop_duplicates()

print('Lenth of orignal df:    ',len(df_filled))
print('Lenght of drop duplicated df:     ',len(df_unique))
print('Row Dropped:      ', len(df_filled)-len(df_unique))
```

```python
df_unique= df_unique.drop_duplicates(subset=['job_title', 'company_name'])

print('Lenth of orignal df:    ',len(df_filled))
print('Lenght of drop duplicated df:     ',len(df_unique))
print('Row Dropped:      ', len(df_filled)-len(df_unique))
```
